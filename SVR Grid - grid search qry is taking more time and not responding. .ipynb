```python
import pandas as pd
df =pd.read_csv('insurance_pre.csv')
```

```python
df
```

```python
df = pd.get_dummies(df,drop_first=True)
df
```

```python
df.columns
```

```python
x_independent = df[['age', 'bmi', 'children',  'sex_male', 'smoker_yes']]
y_depenedent = df[['charges']]
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVR
param_grid = {'kernel':['rbf','poly','sigmoid','linear'],
'C':[10,100,1000,2000,3000],'gamma':['auto','scale']}

grid = GridSearchCV(SVR(), param_grid, refit = True, verbose= 3,n_jobs=-1)
grid.fit(x_independent, y_depenedent)


```

```python
re=grid.cv_results_
print("best params",grid.best_params_)
table=pd.DataFrame.from_dict(re)
```

```python
result=grid.predict([[32,25,2,1,0]])
```
